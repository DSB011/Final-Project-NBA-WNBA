```python

year_list = ['2015','2016','2017','2018','2019']
game_ids = []
home_team_ids = []
away_team_ids =[]
home_team_names = []
away_team_names =[]
dates =[]
home_points = []
away_points = []
arena_names =[]
arena_addresses =[]
arena_cities = []
arena_states = []
arena_zips = []
arena_countries = []

for year in year_list:
    try:
        url = f'http://api.sportradar.us/wnba/trial/v7/en/games/{year}/REG/schedule.json?api_key={your_api_key}' 
        time.sleep(5)
        r = requests.get(url)
        game_stats = r.json()
        games = game_stats['games']
        for i in games:
            game_id = i['id']
            home_team_id = i['home']['id']
            away_team_id = i['away']['id']
            home_team_name = i['home']['name']
            away_team_name = i['away']['name']
            date = i['scheduled']
            if 'home_points' in i:
                home_point = i['home_points']
            else:
                home_point = 'na'
            if 'away_points' in i:
                away_point = i['away_points']
            else:
                away_point = 'na'
            game_ids.append(game_id)
            home_team_ids.append(home_team_id)
            away_team_ids.append(away_team_id)
            home_team_names.append(home_team_name)
            away_team_names.append(away_team_name)
            dates.append(date)
            home_points.append(home_point)
            away_points.append(away_point)
    except:
        print(url)
    stats_dict = {"game_id":game_ids, "home_team_id":home_team_ids,
                  "away_team_id":away_team_ids, "home_team_name":home_team_names,
                  "away_team_name":away_team_names, "date":dates,
                  "home_points":home_points, "away_points":away_points
                 }
   
game_stats_df = pd.DataFrame(stats_dict)
game_stats_df
```
