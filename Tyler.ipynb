```python
id_list = df['Game_id'].tolist()
# id_list = ['0da78f13-73ac-4465-8e31-ecc3029a5dc6','d5393ffd-caf1-4557-84c0-73ec2cfa8629']
attendance_count = []
player_fname = []
player_lname =[]
player_id = []
player_position = []
player_pts = []
player_assist = []
player_reboud = []
free_throw_percent = []
two_pt_percent = []
three_pt_pct = []
steals = []
turnovers = []
game_ids = []
capicity = []
team_played_for = []
team_total_pts = []
home_away = []
data = []
win = []
tech_fouls =[]
flagrant =[]
personal_fouls = []
off_boards = []
min_played = []
game_date =[]


for game_id in id_list:
    time.sleep(10)
    url = f'https://api.sportradar.us/nba/trial/v7/en/games/{game_id}/summary.json?api_key={your_api_key}'
    r = requests.get(url)
    game_stats = r.json()
    game_id = game_stats['id']
    cap = game_stats['venue']['capacity']
    attend = game_stats['attendance']
    team = game_stats['home']['name']
    team_away = game_stats['away']['name']
    total_pts=game_stats['home']['points']
    away_total_pts=game_stats['away']['points']
    home_team = game_stats['home']['players']
    away_team = game_stats['away']['players']
    date = game_stats['scheduled']

    for i in home_team:

        homeplayerfirst = i['first_name']
        homeplayerlast = i['last_name']
        homeplayerposition = i['position']
        homeplayerid = i['id']
        homeplayerpts = i['statistics']['points']
        homeplayerblocks = i['statistics']['blocks']
        homeplayerfreethrowpct = i['statistics']['free_throws_pct']
        homeplayertwoptpct = i['statistics']['two_points_pct']
        homeplayerthreeptpct = i['statistics']['three_points_pct']
        homeplayerturnover = i['statistics']['turnovers']
        homeplayerassist = i['statistics']['assists']
        homeplayerrebounds = i['statistics']['rebounds']
        homeplayersteals = i['statistics']['steals']
        homeplayeroffboards = i['statistics']['offensive_rebounds']
        homeplayertech = i['statistics']['tech_fouls']
        homeplayerflagrant = i['statistics']['flagrant_fouls']
        homeplayerpersonal = i['statistics']['personal_fouls']
        homeplayermin = i['statistics']['minutes']
        player_fname.append(homeplayerfirst)
        player_lname.append(homeplayerlast)
        player_id.append(homeplayerid)
        player_position.append(homeplayerposition)
        player_pts.append(homeplayerpts)
        player_assist.append(homeplayerassist)
        player_reboud.append(homeplayerrebounds)
        free_throw_percent.append(homeplayerfreethrowpct)
        two_pt_percent.append(homeplayertwoptpct)
        three_pt_pct.append(homeplayerthreeptpct)
        steals.append(homeplayersteals)
        turnovers.append(homeplayerturnover)
        game_ids.append(game_id)
        capicity.append(cap)
        home_away.append(1)
        attendance_count.append(attend)
        team_played_for.append(team)
        team_total_pts.append(total_pts)
        tech_fouls.append(homeplayertech)
        flagrant.append(homeplayerflagrant)
        personal_fouls.append(homeplayerpersonal)
        off_boards.append(homeplayeroffboards)
        min_played.append(homeplayermin)
        game_date.append(date)
        if total_pts > away_total_pts:
            win.append(1)
        else:
            win.append(0)
        
    for i in away_team:

        awayplayerfirst = i['first_name']
        awayplayerlast = i['last_name']
        awayplayerposition = i['position']
        awayplayerid = i['id']
        awayplayerpts = i['statistics']['points']
        awayplayerblocks = i['statistics']['blocks']
        awayplayerfreethrowpct = i['statistics']['free_throws_pct']
        awayplayermin = i['statistics']['minutes']
        awayplayertwoptpct = i['statistics']['two_points_pct']
        awayplayerthreeptpct = i['statistics']['three_points_pct']
        awayplayerturnover = i['statistics']['turnovers']
        awayplayerassist = i['statistics']['assists']
        awayplayerrebounds = i['statistics']['rebounds']
        awayplayersteals = i['statistics']['steals']
        awayplayeroffboards = i['statistics']['offensive_rebounds']
        awayplayertech = i['statistics']['tech_fouls']
        awayplayerflagrant = i['statistics']['flagrant_fouls']
        awayplayerpersonal = i['statistics']['personal_fouls']
        player_fname.append(awayplayerfirst)
        player_lname.append(awayplayerlast)
        player_id.append(awayplayerid)
        player_position.append(awayplayerposition)
        player_pts.append(awayplayerpts)
        player_assist.append(awayplayerassist)
        player_reboud.append(awayplayerrebounds)
        free_throw_percent.append(awayplayerfreethrowpct)
        two_pt_percent.append(awayplayertwoptpct)
        three_pt_pct.append(awayplayerthreeptpct)
        steals.append(awayplayersteals)
        turnovers.append(awayplayerturnover)
        game_ids.append(game_id)
        capicity.append(cap)
        home_away.append(0)
        attendance_count.append(attend)
        team_played_for.append(team_away)
        team_total_pts.append(away_total_pts)
        tech_fouls.append(awayplayertech)
        flagrant.append(awayplayerflagrant)
        personal_fouls.append(awayplayerpersonal)
        off_boards.append(awayplayeroffboards)
        min_played.append(awayplayermin)
        game_date.append(date)
        if away_total_pts > total_pts:
            win.append(1)
        else:
            win.append(0)

    stats_dict = {"First_Name":player_fname, "Last_Name":player_lname,
                  "player_id":player_id, "Position":player_position,
                  "Points":player_pts, "Free_Throw_Percent":free_throw_percent,
                  "Two_Pt_Percent": two_pt_percent,
                  "Three_Pt_Percent":three_pt_pct, "Assists":player_assist,
                  "Rebounds":player_reboud, "Offensive_Rebounds": off_boards,
                  "Steals":steals, "Personal_Fouls":personal_fouls,
                  "Flagrant_Fouls":flagrant, "Tech_Fouls":tech_fouls,
                  "Turnovers":turnovers, "Team":team_played_for,
                  "Home_Away":home_away, "win":win, 
                  "Team_points":team_total_pts, "Min_played":min_played, 
                  "Crowd":attendance_count, "Stadium_Cap":capicity,
                  "game_id":game_ids, "game_date":game_date,
                  
                  
                 }
   

player_stats_df = pd.DataFrame(stats_dict)
player_stats_df

```

```python
player_stats_df.to_csv("data_files/player_stats.csv")
```
