```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import bar_chart_race as bcr
import datetime as dt
```

```python
wnba_top10 = pd.read_csv("Data/WBNAfullplayerstatslist.csv")
wnba_top10.head(10)
```

```python
wnba_top10['player_name'] = wnba_top10['First_Name'].str.cat(wnba_top10['Last_Name'], sep =" ")
wnba_top10 = wnba_top10[['player_name', 'Points','Position','game_date']]
wnba_top10
```

```python
wnba_top10['Date'] = pd.to_datetime(wnba_top10['game_date'], errors='coerce')
wnba_top10.dtypes
```

```python
wnba_top10['Date'] = wnba_top10['Date'].dt.normalize()
```

```python
wnba_top10['Date'] = wnba_top10['Date'].dt.date
```

```python
wnba_top10.head()
```

```python
wnba_top_players = wnba_top10.groupby('player_name').sum()['Points']
wnba_top_players = wnba_top_players.sort_values(ascending=False)
wnba_top_players = wnba_top_players.head(10)
wnba_top_players
```

```python
wnba_data = wnba_top10.pivot_table(index='Date',columns='player_name',values='Points')
wnba_data = wnba_data[wnba_top_players.index.values]
wnba_data
```

```python
wnba_data = wnba_data.fillna(0)
wnba_data
```

```python
wnba_data = wnba_data.cumsum()
wnba_data
```

```python
bcr.bar_chart_race(df=wnba_data, filename=None)
```
