```python
import pandas as pd
from datetime import datetime as dt
import numpy as np
import requests
import config
import json
import time
import datetime
from pprint import pprint
```

```python
df = pd.read_csv('data_files/fullplayerstatslist.csv')
```

```python
df
```

```python
df_dropped = df[df['Min_played'] != "00:00"]
df_dropped = df_dropped[df_dropped['Crowd'] != 'Covid']
df_dropped = df_dropped[df_dropped['Crowd'] != '0']
df_dropped= df_dropped[["Points", "Free_Throw_Percent",
                  "Two_Pt_Percent",
                  "Three_Pt_Percent", "Assists",
                  "Rebounds", "Offensive_Rebounds",
                  "Steals", "Personal_Fouls",
                  "Flagrant_Fouls", "Tech_Fouls",
                  "Turnovers",
                  "Home_Away", "win", 
                  "Crowd"
                  ]].reset_index(drop = True)
df_dropped
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X = df_dropped.drop('win', axis=1)
print(X.shape)
```

```python
y = df_dropped['win']
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1)
```

```python
X_train.head()
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model1 = RandomForestClassifier(n_estimators=100)
model1.fit(X_train_scaled, y_train)
```

```python
print(f"Training Data Score: {round(model1.score(X_train_scaled, y_train)*100,2)}%")
print(f"Testing Data Score: {round(model1.score(X_test_scaled, y_test)*100,2)}%")
```

```python
feature_names = X.columns.tolist()
preSelected_features = sorted(zip(model1.feature_importances_, feature_names), reverse=True)
ranked_features = pd.DataFrame(preSelected_features, columns=['Score', 'Feature'])
ranked_features = ranked_features.set_index('Feature')
ranked_features
```
